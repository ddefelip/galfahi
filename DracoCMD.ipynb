```python
import numpy as np
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
from astropy.table import Table
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
draco = SkyCoord.from_name('Draco dSph')
tab = Table.read('Random_position_ddefelip.csv') # should be the surrounding ~30 arcmin around Draco
```

```python
sc = SkyCoord.guess_from_table(tab, unit=u.deg)
tab['dist'] = draco.separation(sc).to(u.arcmin)
```

```python
mtab = tab[tab['dist']<5]
c = mtab['psf_g'] - mtab['psf_i']
m = mtab['psf_r']

plt.scatter(c, m, alpha=.5, s=2)
plt.ylim(23,15)
plt.xlim(-1,4)
```
